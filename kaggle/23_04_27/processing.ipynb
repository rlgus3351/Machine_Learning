```python
plt.figure(figsize=(15,6))
sns.violinplot(data=train, x='race',y='hours-per-week', hue='income',split=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data=train, x='race',y='education-num', hue='income',split=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data=train, x='workclass',y='education-num', hue='income',split=True)
plt.ylim(10,16)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data=train, x='education',y='hours-per-week', hue='income',split=True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(data=train, x='occupation',y='hours-per-week', hue='income',split=True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(train, x='occupation', hue='education-num')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.FacetGrid(train, col = 'race', row = 'sex').map(sns.distplot, 'income')
```
